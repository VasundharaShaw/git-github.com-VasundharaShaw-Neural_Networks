```python
# Shape of the data is very important. PyTorch expects data in a specific shape which is different from other libraries like TensorFlow/Keras. 
# It is important to understand how to reshape data correctly. 
```

```python
inputs = [1,2,3,2.5] ## 4 inputs to the layer

weights = [[0.2,0.8,-0.5,1.0]] ## All weights in one list

biases = 2 ## All biases in one list

output = np.dot(weights, inputs) + biases ## Using numpy for matrix multiplication and adding biases. We use weights first to get correct shape of output because numpy follows standard matrix multiplication rules.

print(output) ## Print outputs of all neurons
```
